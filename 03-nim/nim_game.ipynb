# NIM game
**Rules** : https://en.wikipedia.org/wiki/Nim

```python
s1 = ['I','III','I','III','H']
```

```python
def get_substates(state):
  substates = []
  for i in range(0,len(state)-1):
    if state[i] == '':
      continue
    for j in range(0,len(state[i])):
      new_state = state.copy()
      new_state[i] = "I"*j
      new_state[-1] = "C" if new_state[-1] == "H" else "H"
      substates.append(new_state)
  return substates
```

```python
for s in get_substates(s1):
  print(s)
```

```python
def check_over(state):
  if state[0] == "" and state[1] == "" and state[2] == "" and state[3] == "":
    return (True,state[-1])
  return (False,"")
```

```python
def print_tree(state, indent=0):
  print("  "*indent,state)
  result = check_over(state)
  if result[0]:
    print("  "*indent, result[1], " Wins!")
    return
  for s2 in get_substates(state):
    print_tree(s2,indent+1)
```

```python
print_tree(s1)
```
